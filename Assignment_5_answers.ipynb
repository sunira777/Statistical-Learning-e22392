## Part A: Theoretical Fundamentals (Maximum Likelihood & Decision Space)

1]

The raw Maximum Likelihood Estimator for the covariance matrix is defined as:$$\hat{\Sigma}_{\text{MLE}} = \frac{1}{n}\sum_{i=1}^{n}(X_i - \hat{\mu}_n)(X_i - \hat{\mu}_n)^T$$We can expand the term inside the summation by adding and subtracting the true population mean $\mu$:$$X_i - \hat{\mu}_n = (X_i - \mu) - (\hat{\mu}_n - \mu)$$Taking the outer product and summing over $i$ yields:$$\sum_{i=1}^{n}(X_i - \hat{\mu}_n)(X_i - \hat{\mu}_n)^T = \sum_{i=1}^{n}(X_i - \mu)(X_i - \mu)^T - \sum_{i=1}^{n}(X_i - \mu)(\hat{\mu}_n - \mu)^T - \sum_{i=1}^{n}(\hat{\mu}_n - \mu)(X_i - \mu)^T + \sum_{i=1}^{n}(\hat{\mu}_n - \mu)(\hat{\mu}_n - \mu)^T$$Noting that $\sum_{i=1}^{n}(X_i - \mu) = n(\hat{\mu}_n - \mu)$, the two middle cross-product terms each become $-n(\hat{\mu}_n - \mu)(\hat{\mu}_n - \mu)^T$. The final term sums to $+n(\hat{\mu}_n - \mu)(\hat{\mu}_n - \mu)^T$. Combining these gives:$$\sum_{i=1}^{n}(X_i - \hat{\mu}_n)(X_i - \hat{\mu}_n)^T = \sum_{i=1}^{n}(X_i - \mu)(X_i - \mu)^T - n(\hat{\mu}_n - \mu)(\hat{\mu}_n - \mu)^T$$Now, take the mathematical expectation of both sides. By the definition of population variance, $E[(X_i - \mu)(X_i - \mu)^T] = \Sigma$. Since the samples are $i.i.d.$, the variance of the sample mean is $E[(\hat{\mu}_n - \mu)(\hat{\mu}_n - \mu)^T] = \frac{1}{n}\Sigma$. Substituting these properties:$$E\left[\sum_{i=1}^{n}(X_i - \hat{\mu}_n)(X_i - \hat{\mu}_n)^T\right] = \sum_{i=1}^{n}\Sigma - n\left(\frac{1}{n}\Sigma\right) = n\Sigma - \Sigma = (n - 1)\Sigma$$Dividing by the sample size $n$ to find the expectation of $\hat{\Sigma}_{\text{MLE}}$:$$E[\hat{\Sigma}_{\text{MLE}}] = \frac{1}{n} E\left[\sum_{i=1}^{n}(X_i - \hat{\mu}_n)(X_i - \hat{\mu}_n)^T\right] = \frac{n-1}{n}\Sigma$$This proves that $\hat{\Sigma}_{\text{MLE}}$ is downwardly biased in finite samples. To eliminate this bias and yield the unique unbiased estimator $S$, we apply Bessel's Correction by multiplying by $\frac{n}{n-1}$:$$S = \frac{n}{n-1}\hat{\Sigma}_{\text{MLE}} = \frac{1}{n-1}\sum_{i=1}^{n}(X_i - \hat{\mu}_n)(X_i - \hat{\mu}_n)^T \implies E[S] = \Sigma$$

2]

*   Type I Error ($\alpha$): The probability of a False Alarm. In SHM, this occurs when the diagnostic layer rejects the null hypothesis $H_0$ (concluding the asset is damaged) when the structural asset is actually completely healthy.

*   Type II Error ($\beta$): The probability of a Missed Detection. In SHM, this occurs when the diagnostic layer fails to reject $H_0$ (concluding the asset is healthy) when the structure has actually developed a critical flaw or drift.

3] Engineering Consequences of an Ultra-Conservative Threshold ($\alpha = 0.0001$)



*   Impact on Statistical Power: Lowering $\alpha$ to an ultra-conservative level forces the critical threshold value deep into the tail of the null distribution. Because Type I and Type II errors are inversely coupled for a fixed sample size, minimizing $\alpha$ causes $\beta$ to spike dramatically. The statistical power of the diagnostic layer, defined as $\pi = 1 - \beta$, drops precipitously, rendering the system practically blind to early, low-magnitude structural drifts.
*  Confidence Ellipsoid Geometry: The healthy operation region is defined by the confidence ellipsoid centered at the baseline mean vector. Mathematically, its geometric volume is determined by the Mahalanobis distance boundary threshold $c_{\alpha}$. Setting $\alpha = 0.0001$ vastly increases $c_{\alpha}$, causing the geometric volume of the "healthy operation" confidence ellipsoid to expand significantly to encompass almost all historical and extreme stochastic noise variations.





## Part B: Theoretical Extension (Asymptotic Distributions & Slutsky's Theorem)

1]

Let $X_n$ and $Y_n$ be sequences of random vectors/matrices. If $X_n$ converges in distribution to a random variable $X$ ($X_n \xrightarrow{d} X$), and $Y_n$ converges in probability to a deterministic constant $c$ ($Y_n \xrightarrow{p} c$), then:$X_n + Y_n \xrightarrow{d} X + c$$Y_n X_n \xrightarrow{d} cX$$Y_n^{-1} X_n \xrightarrow{d} c^{-1}X \quad (\text{provided } c \neq 0)$



2]

From the Lindeberg-Lévy Multivariate CLT, as $n \to \infty$:$$\sqrt{n}(\hat{\mu}_n - \mu) \xrightarrow{d} N(0, \Sigma) \implies \Sigma^{-1/2} \sqrt{n}(\hat{\mu}_n - \mu) \xrightarrow{d} N(0, I_m)$$By the Weak Law of Large Numbers (WLLN), the empirical sample covariance matrix $S$ is a consistent estimator of the population covariance matrix, meaning $S \xrightarrow{p} \Sigma$ as $n \to \infty$. Because matrix inversion and the matrix square root are continuous operations, the continuous mapping theorem dictates that:$$S^{-1/2} \xrightarrow{p} \Sigma^{-1/2}$$Applying Slutsky's Theorem (specifically case 2 for products), we can scale our CLT convergence sequence by substituting our consistent probability limit:$$S^{-1/2} \sqrt{n}(\hat{\mu}_n - \mu) = \left(S^{-1/2} \Sigma^{1/2}\right) \cdot \left(\Sigma^{-1/2} \sqrt{n}(\hat{\mu}_n - \mu)\right) \xrightarrow{d} I_m \cdot N(0, I_m) = N(0, I_m)$$Squaring the operations back out to express the asymptotic distribution of the empirical mean vector yields:$$\sqrt{n}(\hat{\mu}_n - \mu) \xrightarrow{d} N(0, S) \implies \hat{\mu}_n \sim N\left(\mu, \frac{1}{n}S\right)$$This fundamentally justifies using the sample covariance $S$ in place of the unobservable parameter $\Sigma$ for finite but large engineered snapshots.

## Part C: Numerical Verification (MANOVA & Bartlett's $\chi^2$)

```python
import numpy as np
import pandas as pd
import scipy.stats as stats

# --- DO NOT MODIFY: Synthetic Strain Sensor Data Generator ---
np.random.seed(42)
n_samples = 5000
n_features = 3

base_data = np.random.multivariate_normal(
    mean=[0.5, -0.2, 1.1],
    cov=[[0.09, 0.02, 0.01],
         [0.02, 0.06, 0.03],
         [0.01, 0.03, 0.05]],
    size=n_samples
)

# Injecting a subtle, hidden structural drift in the final 1,000 snapshots
base_data[4000:, 0] += 0.015  # Sensor 1 drift
base_data[4000:, 2] -= 0.010  # Sensor 3 drift

df_strain = pd.DataFrame(base_data, columns=['Strain_Ch1', 'Strain_Ch2', 'Strain_Ch3'])
# ----------------------------------------------------------------

def verify_first_moment_homogeneity(df: pd.DataFrame, g_chunks: int = 5) -> dict:
    """
    Partitions the dataset into g chunks and evaluates first-moment homogeneity
    via Wilks' Lambda and Bartlett's Chi-Square asymptotic transformation.
    """
    X = df.values
    n, m = X.shape
    chunk_size = n // g_chunks

    # 2. Compute global mean
    global_mean = np.mean(X, axis=0)

    # Initialize matrices W and B
    W = np.zeros((m, m))
    B = np.zeros((m, m))

    # 1 & 3. Process individual chunks
    for j in range(g_chunks):
        chunk = X[j * chunk_size : (j + 1) * chunk_size, :]
        chunk_mean = np.mean(chunk, axis=0)

        # Within-chunk scatter contribution
        chunk_centered = chunk - chunk_mean
        W += chunk_centered.T @ chunk_centered

        # Between-chunk scatter contribution
        mean_diff = (chunk_mean - global_mean).reshape(-1, 1)
        B += chunk_size * (mean_diff @ mean_diff.T)

    # 4. Compute Wilks' Lambda
    det_W = np.linalg.det(W)
    det_B_plus_W = np.linalg.det(B + W)
    wilks_lambda = det_W / det_B_plus_W

    # Bartlett's Chi-Square transformation
    df_chi = m * (g_chunks - 1)
    bartlett_multiplier = -(n - 1 - (m + g_chunks) / 2.0)
    chi2_calc = bartlett_multiplier * np.log(wilks_lambda)

    # Calculate p-value
    p_value = 1.0 - stats.chi2.cdf(chi2_calc, df_chi)

    # Evaluation conclusion rule
    alpha = 0.05
    conclusion = "Rejected H0: First moment is heterogeneous (Baseline center shifted)." if p_value <= alpha \
                 else "Fail to Reject H0: First moment is homogeneous."

    return {
        "Wilks' Lambda": float(wilks_lambda),
        "Bartlett Chi2 Statistic": float(chi2_calc),
        "Degrees of Freedom": int(df_chi),
        "p-value": float(p_value),
        "Conclusion": conclusion
    }

# Execute completed function
results = verify_first_moment_homogeneity(df_strain, g_chunks=5)
for k, v in results.items():
    print(f"{k}: {v if isinstance(v, str) else f'{v:.6f}'}")
```

# 2. Geometric Subspace Optimization via PCA

## Part A: Theoretical Fundamentals

1]

Given $Z_i = P^T \tilde{X}_i$ and $\Sigma = E[\tilde{X}_i \tilde{X}_i^T] = P \Lambda P^T$, we find the covariance matrix of the projected space:$$E[Z_i Z_i^T] = E[(P^T \tilde{X}_i)(P^T \tilde{X}_i)^T]$$Using the linear algebra transpose property $(AB)^T = B^T A^T$:$$E[Z_i Z_i^T] = E[P^T \tilde{X}_i \tilde{X}_i^T P]$$Because the projection matrix $P$ is derived deterministically via the eigenvectors of the baseline network, it is a constant matrix under the expectation operator:$$E[Z_i Z_i^T] = P^T E[\tilde{X}_i \tilde{X}_i^T] P = P^T \Sigma P$$Substitute the spectral decomposition of $\Sigma$:$$E[Z_i Z_i^T] = P^T (P \Lambda P^T) P$$Because $P$ is an orthogonal matrix, its columns satisfy $P^T P = I$. Grouping the operations:$$E[Z_i Z_i^T] = (P^T P) \Lambda (P^T P) = I \Lambda I = \Lambda$$Geometric and Statistical Meaning: The resulting diagonal matrix $\Lambda$ proves that the cross-covariances between distinct coordinate features $Z_{i,j}$ and $Z_{i,k}$ ($j \neq k$) are exactly zero. Geometrically, PCA rotates the coordinate frames to align with the orthogonal axes of maximum variance, meaning all linear correlations between channels are decoupled.

2]
We utilize the cyclic property of the matrix trace ($\text{tr}(ABC) = \text{tr}(CAB)$):$$\text{Total Variance in Principal Space} = \text{tr}(\Lambda)$$From our diagonalization proof, we know $\Lambda = P^T \Sigma P$. Substituting this configuration:$$\text{tr}(\Lambda) = \text{tr}(P^T \Sigma P)$$Applying the cyclic permutation property where $A = P^T \Sigma$ and $B = P$:$$\text{tr}((P^T \Sigma) P) = \text{tr}(P (P^T \Sigma))$$Since $P P^T = I$ due to structural matrix orthogonality:$$\text{tr}(P P^T \Sigma) = \text{tr}(I \Sigma) = \text{tr}(\Sigma) = \sum_{j=1}^{m} \lambda_j$$This confirms that the global information (total variance) is preserved exactly during linear coordinate transformations.

### 3. Subspace Formulations, Pythagorean Identity & Diagnostic Matching



*   Cumulative Explained Variance Ratio $\Phi(k)$: $\Phi(k) = \frac{\sum_{j=1}^{k} \hat{\lambda}_j}{\sum_{j=1}^{m} \hat{\lambda}_j}$
*   Residual Unexplained Variance Ratio $\Psi(k)$: $\Psi(k) = 1 - \Phi(k) = \frac{\sum_{j=k+1}^{m} \hat{\lambda}_j}{\sum_{j=1}^{m} \hat{\lambda}_j}$
*  Pythagorean Subspace Decomposition Proof: The centered realization is partitioned into orthogonal projections: $(x_i - \hat{\mu}_n) = \hat{P}_k z_{i,k} + \hat{P}_{m-k} z_{i,m-k}$. Taking the squared Euclidean norm:$$\|x_i - \hat{\mu}_n\|^2 = \left(\hat{P}_k z_{i,k} + \hat{P}_{m-k} z_{i,m-k}\right)^T \left(\hat{P}_k z_{i,k} + \hat{P}_{m-k} z_{i,m-k}\right)$$Expanding this matrix product:$$\|x_i - \hat{\mu}_n\|^2 = z_{i,k}^T \hat{P}_k^T \hat{P}_k z_{i,k} + 2 z_{i,k}^T \hat{P}_k^T \hat{P}_{m-k} z_{i,m-k} + z_{i,m-k}^T \hat{P}_{m-k}^T \hat{P}_{m-k} z_{i,m-k}$$Because the sub-blocks are mutually orthogonal eigenvectors, $\hat{P}_k^T \hat{P}_k = I_k$, $\hat{P}_{m-k}^T \hat{P}_{m-k} = I_{m-k}$, and the cross-product matrix $\hat{P}_k^T \hat{P}_{m-k} = 0$. The cross term completely vanishes, simplifying exactly to:$$\|x_i - \hat{\mu}_n\|^2 = z_{i,k}^T z_{i,k} + z_{i,m-k}^T z_{i,m-k} = \|z_{i,k}\|^2 + \|z_{i,m-k}\|^2$$
*  Diagnostic Matching:

Extreme Operational Wind Load $\implies$ Hotelling's $T^2$ Statistic. Operational loads vary the correlation state within the established physical subspace. The structure is intact, but the energy shifts along normal operational paths, increasing $T^2$ values while residuals remain low.

Internal Structural Fatigue Crack $\implies$ Residual $Q$ Statistic (SPE). A physical fracture alters the physical load pathway of the asset. Because the underlying structural transfer function changes, the input signals can no longer be reconstructed by the nominal baseline eigenvectors. This forces data energy to project out of the explained subspace into the residual layer, causing a significant spike in the $Q$ statistic.







## Part B: PCA Numerical Optimization Pipeline

```python
import numpy as np

def optimize_pca_subspace(X_raw: np.ndarray) -> dict:
    """
    Standardizes input matrix, extracts sorted eigenvalues, and calculates
    the empirical means of Hotelling's T2 and Q metrics across k cutting boundaries.
    """
    n, m = X_raw.shape
    # 1. Standardize and center realizations
    X_centered = X_raw - np.mean(X_raw, axis=0)
    X_scaled = X_centered / np.std(X_centered, axis=0, ddof=1)

    # 2. Decompose sample covariance matrix
    S = np.cov(X_scaled, rowvar=False)
    eigenvalues, eigenvectors = np.linalg.eigh(S)

    # Sort in descending order
    idx = np.argsort(eigenvalues)[::-1]
    lambdas = eigenvalues[idx]
    P = eigenvectors[:, idx]

    # Project scores
    Z = X_scaled @ P

    metrics = {}
    # 3. Loop through subspace cutoff dimensions
    for k in [1, 2, 3]:
        # Hotelling's T2 calculation per sample
        t2_samples = np.sum((Z[:, :k] ** 2) / lambdas[:k], axis=1)

        # Q-statistic (SPE) calculation per sample
        q_samples = np.sum(Z[:, k:] ** 2, axis=1)

        # 4. Compute expected values across all rows
        metrics[f"k={k}"] = {
            "Mean_T2": float(np.mean(t2_samples)),
            "Mean_Q": float(np.mean(q_samples))
        }
    return metrics

# Execution demonstration with nominal mock data matrix (n=3000, m=4)
np.random.seed(42)
mock_data = np.random.normal(0, 1, (3000, 4))
pca_optimization_results = optimize_pca_subspace(mock_data)
print("PCA Subspace Optimization Metrics:")
print(pca_optimization_results)
```

# 3. Latent Subspace Decomposition via Factor Analysis (FA)

## Part A: Theoretical Fundamentals

1]Given $Z_i = \Lambda F_i + \epsilon_i$, we expand the definition of the correlation matrix $R$:$$R = E[Z_i Z_i^T] = E[(\Lambda F_i + \epsilon_i)(\Lambda F_i + \epsilon_i)^T]$$Distributing the matrix transpose and multiplication operations:$$R = E\left[\Lambda F_i F_i^T \Lambda^T + \Lambda F_i \epsilon_i^T + \epsilon_i F_i^T \Lambda^T + \epsilon_i \epsilon_i^T\right]$$Applying the linear property of expectations and pulling deterministic matrices out of the terms:$$R = \Lambda E[F_i F_i^T] \Lambda^T + \Lambda E[F_i \epsilon_i^T] + E[\epsilon_i F_i^T] \Lambda^T + E[\epsilon_i \epsilon_i^T]$$Substituting our strict orthogonal structural constraints ($E[F_i F_i^T] = I_{k \times k}$, $E[\epsilon_i F_i^T] = 0$, and $E[\epsilon_i \epsilon_i^T] = \Psi$):$$R = \Lambda (I) \Lambda^T + \Lambda (0) + (0) \Lambda^T + \Psi \implies R = \Lambda \Lambda^T + \Psi$$

*  Communality ($h_j^2$): The portion of a sensor's variance driven by the shared common structural factors. In asset diagnostics, high communality indicates the sensor is successfully monitoring global physical states (e.g., system load, primary vibrations).
*  Uniqueness ($\phi_j^2$): The portion of variance completely isolated to that specific sensor channel. Physically, a spike in uniqueness isolated to a single channel reveals localized electrical instrument noise or hardware degradation rather than an actual structural failure.

### 2]About Rotations and simple structure

*  Conceptual Difference: A raw PCA loading vector forces maximum variance extraction into a strict orthogonal hierarchy, which mixes features across all channels. A Varimax-rotated factor loading vector shifts the axes to polarize the weights—making them close to $1.0$ or $0.0$ on individual components. This achieves a "simple structure" that links specific sensors to specific physical phenomena.
*  Invariance Proof: Let $T$ be an orthogonal rotation matrix such that $T T^T = I$, and define the rotated loading matrix as $\Lambda^* = \Lambda T$. Substituting this back into the covariance approximation:$$R \approx \Lambda^* (\Lambda^*)^T + \Psi = (\Lambda T)(\Lambda T)^T + \Psi = \Lambda (T T^T) \Lambda^T + \Psi = \Lambda I \Lambda^T + \Psi = \Lambda \Lambda^T + \Psi$$Because the rotation matrix cancels out internally, the individual sensor communalities ($h_j^2$, the diagonal elements of $\Lambda \Lambda^T$) and the global matrix approximation remain completely unchanged.



3]Derive MMSE

We construct the joint vector

$Y_i = [Z_i^T, F_i^T]^T$. The cross-covariance blocks are:$$E[F_i Z_i^T] = E[F_i (\Lambda F_i + \epsilon_i)^T] =
E[F_i F_i^T] \Lambda^T + E[F_i \epsilon_i^T] = I \Lambda^T + 0 =\Lambda^T$$Thus, the full joint population covariance matrix expands exactly to:$$\Sigma_{YY} = \begin{bmatrix} R & \Lambda \\ \Lambda^T & I_{k \times k} \end{bmatrix}$$Under multivariate normality, the conditional expectation identity dictates that:$$E[X_2 | x_1] = \mu_2 + \Sigma_{21}\Sigma_{11}^{-1}(x_1 - \mu_1)$$Mapping our factor blocks where $X_2 = F_i$ (latent factor), $X_1 = Z_i$ (observed vector), $\mu_1 = 0$, $\mu_2 = 0$, $\Sigma_{21} = \Lambda^T$, and $\Sigma_{11} = R$:$$f_i = E[F_i | Z_i] = 0 + \Lambda^T R^{-1}(Z_i - 0) \implies f_i = \Lambda^T R^{-1} Z_i$$

## Part B: Factor Partition Engine & UI Dashboard

```python
import numpy as np
import pandas as pd
from sklearn.decomposition import FactorAnalysis
import plotly.subplots as sp
import plotly.graph_objects as go

# --- DO NOT MODIFY: Synthetic Asset Grid Generation Data ---
np.random.seed(42)
n_samples = 2500

f1 = np.random.normal(0, 1, n_samples)  # Primary structural mode
f2 = np.random.normal(0, 1, n_samples)  # Secondary operational mode

s1 = 0.85 * f1 + 0.10 * f2 + np.random.normal(0, 0.3, n_samples)
s2 = 0.80 * f1 + 0.15 * f2 + np.random.normal(0, 0.35, n_samples)
s3 = 0.12 * f1 + 0.90 * f2 + np.random.normal(0, 0.25, n_samples)
s4 = 0.02 * f1 + 0.05 * f2 + np.random.normal(0, 1.40, n_samples)

df_asset = pd.DataFrame(
    data=np.vstack([s1, s2, s3, s4]).T,
    columns=['Sensor_1', 'Sensor_2', 'Sensor_3', 'Sensor_4']
)
# ----------------------------------------------------------------

def run_factor_diagnostics_pipeline(df: pd.DataFrame, k_factors: int = 2):
    n, m = df.shape

    # 1. Dimensionality Constraint Layer Guardrails
    if n <= m:
        raise ValueError(f"Processing Rejected: Snapshot count n ({n}) must exceed channel count m ({m}).")
    if k_factors >= m:
        raise ValueError(f"Configuration Error: Latent features k ({k_factors}) must be strictly less than m ({m}).")

    # Z-Score Mapping Layer with Guardrails
    feature_names = df.columns.tolist()
    Z = np.zeros((n, m))
    for j, col in enumerate(feature_names):
        std_j = df[col].std(ddof=1)
        if std_j < 1e-15:
            std_j = 1e-15  # Zero-variance cap guardrail
        Z[:, j] = (df[col].values - df[col].mean()) / std_j

    # 2. Information Decomposition Layer via Varimax FA
    fa = FactorAnalysis(n_components=k_factors, rotation='varimax', random_state=42)
    fa.fit(Z)

    # Extract structural metrics
    Lambda_rotated = fa.components_.T  # Shape: (m, k)
    uniqueness = fa.noise_variance_     # Shape: (m,)
    communality = 1.0 - uniqueness      # Shape: (m,)

    # 3. Latent Score Projection via Thomson's Method
    R_empirical = np.corrcoef(Z, rowvar=False)
    R_inv = np.linalg.inv(R_empirical)
    # Factor Score Transformation Matrix: Lambda^T @ R^-1
    thomson_mapping_weight = Lambda_rotated.T @ R_inv
    factor_scores = Z @ thomson_mapping_weight.T

    # Performance Diagnostic Telemetry
    avg_communality = np.mean(communality) * 100
    avg_uniqueness = np.mean(uniqueness) * 100
    print(f"=== SYSTEM PERFORMANCE DIAGNOSTIC TELEMETRY ===")
    print(f"Average System Communality %: {avg_communality:.2f}%")
    print(f"Average System Uniqueness %:  {avg_uniqueness:.2f}%\n")

    # 4. Interface Layout Engineering (2x2 Grid Layout)
    fig = sp.make_subplots(
        rows=2, cols=2,
        horizontal_spacing=0.24,
        vertical_spacing=0.28,
        subplot_titles=(
            "Structural Loadings Matrix Heatmap",
            "Variance Partitioning Profile",
            "Sensor Uniqueness Line Profile",
            "Latent Factor Variance Distribution"
        )
    )

    # Subplot(1,1): Structural Loadings Heatmap
    heatmap_trace = go.Heatmap(
        z=np.abs(Lambda_rotated),
        x=[f"Factor {r+1}" for r in range(k_factors)],
        y=feature_names,
        colorscale='YlOrRd',
        colorbar=dict(
            x=-0.15, len=0.38, y=0.78,
            yanchor="middle", xanchor="right",
            title="Abs Loading"
        ),
        showscale=True
    )
    fig.add_trace(heatmap_trace, row=1, col=1)
    fig.update_yaxes(autorange="reversed", row=1, col=1)

    # Subplot(1,2): Variance Partitioning Profile (Stacked Horizontal Bar)
    fig.add_trace(go.Bar(
        y=feature_names, x=communality * 100,
        name="Shared Communality (h²)", orientation='h',
        marker=dict(color='#1f77b4')
    ), row=1, col=2)
    fig.add_trace(go.Bar(
        y=feature_names, x=uniqueness * 100,
        name="Isolated Uniqueness (φ²)", orientation='h',
        marker=dict(color='#ff7f0e')
    ), row=1, col=2)
    fig.update_xaxes(range=[0, 100], row=1, col=2)
    fig.update_layout(barmode='stack')

    # Subplot(2,1): Sensor Uniqueness Line Profile
    fig.add_trace(go.Scatter(
        x=feature_names, y=uniqueness,
        mode='lines+markers', name='Noise Floor Threshold',
        line=dict(color='#d62728', width=2, dash='dashdot'),
        marker=dict(symbol='x', size=8)
    ), row=2, col=1)
    fig.update_xaxes(tickangle=25, row=2, col=1)

    # Subplot(2,2): Latent Factor Variance Distribution
    factor_variances = np.var(factor_scores, axis=0, ddof=1)
    fig.add_trace(go.Bar(
        x=[f"Factor {r+1}" for r in range(k_factors)], y=factor_variances,
        name='Factor Score Variance',
        marker=dict(color='#2ca02c', line=dict(color='black', width=0.5))
    ), row=2, col=2)

    # Global Dashboard Canvas Adjustments
    fig.update_layout(
        template='plotly_white',
        width=1250, height=750,
        margin=dict(t=150, b=60, l=140, r=80),
        legend=dict(
            orientation="h",
            x=0.5, y=1.02,
            xanchor="center", yanchor="bottom"
        )
    )

    fig.show()

# Run the unified engineering execution code
run_factor_diagnostics_pipeline(df_asset, k_factors=2)
```

# Analytical case study answer

* Q1


1.   Physical Nature of Sensor_4: A uniqueness value exceeding $98\%$ indicates that the structural data streams entering this channel contain almost zero shared structural information. Statistically, Sensor_4 is entirely decoupled from the global physical phenomena driving the asset; physically, this identifies an isolated instrumentation fault, ungrounded terminal connection, or severe localized electrical white noise interference.
2.   How Noise Tricks the PCA Engine: PCA works by maximizing global variance without distinguishing its source. Because Sensor_4 has a large noise scale ($\sigma^2 \approx 2.0$), its total absolute variance is much larger than that of the clean structural channels. PCA treats this noise as a dominant feature, aligning a major principal component directly with the noise axis of Sensor_4 and inflating its top eigenvalues.
3.  Automation Framework Risk: Relying only on PCA creates a major false alarm and blind spot vulnerability. If Sensor_4 suffers an electrical spike or additional noise drift, the PCA engine will misinterpret it as a massive global plant anomaly, triggering a costly, unnecessary shutdown. Conversely, a real structural cracking failure could be hidden because its energy scale is smaller than the noise on Sensor_4.


*   Q2.
1. How Varimax Relaxes the Hierarchy Rule: Traditional PCA enforces a strict mathematical hierarchy where the first component aims to capture as much mixed variance as possible, leading to messy, uninterpretable eigenvector profiles. Factor Analysis with a Varimax rotation drops this global variance maximization rule. It rotates the factor space axes to maximize the variance of the squared loadings within each column, forcing individual weights to drive toward $1.0$ or $0.0$. This separates mixed variable patterns into clean, distinct blocks.

2.  Operator Troubleshooting Advantage: If a structural failure occurs, an unrotated PCA matrix displays complex cross-coupled coefficients across all channels, making it difficult to isolate the source of the issue. The rotated FA loading matrix maps directly to independent physical sub-systems. An operator can instantly see that Factor 1 represents the primary structural load pathway (Sensors 1 & 2) while Factor 2 represents the secondary operational mode (Sensor 3). This lets them locate the physical failure without having to parse complex linear algebra mixtures.


*  Q3.
1. Justification using Uniqueness ($\phi^2$) Metrics:In an actual plant deployment, tracking raw PCA projections leaves your $T^2$ and $Q$ bounds vulnerable to single-point hardware sensor faults. If a single hardware channel suffers an electrical short or calibration loss, its correlation with the rest of the multi-sensor grid instantly drops to zero. The FA engine isolates this failure by absorbing the anomalous variance into that channel's individual uniqueness metric ($\phi_4^2 \to 1.0$) while keeping the structural loading matrix $\Lambda$ stable.
