# Question 1 - Statistical Mechanics of the Quantum Harmonic Oscillator

## Partition Function

The canonical partition function is

$$
Z(T)=\sum_{j=0}^{\infty}e^{-E_j/(k_BT)}
$$

Substituting

$$
E_j=\hbar\omega\left(j+\frac12\right),
$$

gives

$$
Z(T)=\sum_{j=0}^{\infty}
e^{-\frac{\hbar\omega(j+\frac12)}{k_BT}}.
$$

Factor out the constant term,

$$
Z(T)=
e^{-\frac{\hbar\omega}{2k_BT}}
\sum_{j=0}^{\infty}
\left(e^{-\frac{\hbar\omega}{k_BT}}\right)^j.
$$

Since

$$
0<e^{-\frac{\hbar\omega}{k_BT}}<1,
$$

the series is geometric with common ratio

$$
r=e^{-\frac{\hbar\omega}{k_BT}},
$$

therefore

$$
\sum_{j=0}^{\infty}r^j=\frac1{1-r}.
$$

Hence,

$$
\boxed{
Z(T)=
\frac{
e^{-\frac{\hbar\omega}{2k_BT}}
}{
1-e^{-\frac{\hbar\omega}{k_BT}}
}
}
$$

#2. Thermodynamic PotentialsA. Helmholtz Free Energy $F(T)$Using the definition $F(T) = -k_B T \ln Z$:
$$F(T) = -k_B T \ln \left( \frac{e^{-\frac{\hbar\omega}{2k_B T}}}{1 - e^{-\frac{\hbar\omega}{k_B T}}} \right)$$Applying log properties ($\ln\frac{a}{b} = \ln a - \ln b$):$$F(T) = -k_B T \left[ -\frac{\hbar\omega}{2k_B T} - \ln\left(1 - e^{-\frac{\hbar\omega}{k_B T}}\right) \right] = \frac{\hbar\omega}{2} + k_B T \ln\left(1 - e^{-\frac{\hbar\omega}{k_B T}}\right)$$

B. Internal Energy $U(T)$Expressing $\ln Z$ strictly in terms of $\beta$:$$\ln Z = -\frac{1}{2}\beta\hbar\omega - \ln\left(1 - e^{-\beta\hbar\omega}\right)$$Taking the negative partial derivative with respect to $\beta$:$$U(T) = -\frac{\partial \ln Z}{\partial \beta} = -\left[ -\frac{\hbar\omega}{2} - \frac{1}{1 - e^{-\beta\hbar\omega}} \cdot \left(-e^{-\beta\hbar\omega}\right) \cdot (-\hbar\omega) \right]$$$$U(T) = \frac{\hbar\omega}{2} + \frac{\hbar\omega e^{-\beta\hbar\omega}}{1 - e^{-\beta\hbar\omega}}$$Dividing the numerator and denominator of the second term by $e^{-\beta\hbar\omega}$ and restoring $\beta = \frac{1}{k_B T}$:$$U(T) = \hbar\omega \left( \frac{1}{2} + \frac{1}{e^{\frac{\hbar\omega}{k_B T}} - 1} \right)$$

#3. Entropy $S(T)$Using the relation $S = -\left(\frac{\partial F}{\partial T}\right)$:

$$S(T) = -\frac{\partial}{\partial T} \left[ \frac{\hbar\omega}{2} + k_B T \ln\left(1 - e^{-\frac{\hbar\omega}{k_B T}}\right) \right]$$Applying the product rule to the second term:$$S(T) = -k_B \ln\left(1 - e^{-\frac{\hbar\omega}{k_B T}}\right) - k_B T \cdot \frac{1}{1 - e^{-\frac{\hbar\omega}{k_B T}}} \cdot \left(-e^{-\frac{\hbar\omega}{k_B T}}\right) \cdot \left(-\frac{\hbar\omega}{k_B T^2}\right)$$$$S(T) = -k_B \ln\left(1 - e^{-\frac{\hbar\omega}{k_B T}}\right) + \frac{\hbar\omega}{T} \frac{1}{e^{\frac{\hbar\omega}{k_B T}} - 1}$$Verification of $U = F + TS$Substituting our expressions for $F(T)$ and $S(T)$ into the right side:$$F + TS = \left[\frac{\hbar\omega}{2} + k_B T \ln\left(1 - e^{-\frac{\hbar\omega}{k_B T}}\right)\right] + T \left[-k_B \ln\left(1 - e^{-\frac{\hbar\omega}{k_B T}}\right) + \frac{\hbar\omega}{T} \frac{1}{e^{\frac{\hbar\omega}{k_B T}} - 1}\right]$$The logarithmic boundary terms cancel out identically:$$F + TS = \frac{\hbar\omega}{2} + \frac{\hbar\omega}{e^{\frac{\hbar\omega}{k_B T}} - 1} = \hbar\omega \left( \frac{1}{2} + \frac{1}{e^{\frac{\hbar\omega}{k_B T}} - 1} \right) = U(T)$$The thermodynamic consistency holds.

#4. Heat Capacity and Hyperbolic FormA. Direct Derivative Calculation$$C_V = \frac{\partial U}{\partial T} = \frac{\partial}{\partial T} \left[ \frac{\hbar\omega}{2} + \hbar\omega \left(e^{\frac{\hbar\omega}{k_B T}} - 1\right)^{-1} \right]
$$Using the chain rule:$$C_V = \hbar\omega \cdot (-1)\left(e^{\frac{\hbar\omega}{k_B T}} - 1\right)^{-2} \cdot e^{\frac{\hbar\omega}{k_B T}} \cdot \left(-\frac{\hbar\omega}{k_B T^2}\right) = k_B \left(\frac{\hbar\omega}{k_B T}\right)^2 \frac{e^{\frac{\hbar\omega}{k_B T}}}{\left(e^{\frac{\hbar\omega}{k_B T}} - 1\right)^2}$$B. Hyperbolic ProofMultiply the numerator and denominator by $e^{-\frac{\hbar\omega}{k_B T}}$:$$C_V = k_B \left(\frac{\hbar\omega}{k_B T}\right)^2 \frac{1}{e^{-\frac{\hbar\omega}{k_B T}}\left(e^{\frac{\hbar\omega}{k_B T}} - 1\right)^2} = k_B \left(\frac{\hbar\omega}{k_B T}\right)^2 \frac{1}{\left(e^{\frac{\hbar\omega}{2k_B T}} - e^{-\frac{\hbar\omega}{2k_B T}}\right)^2}$$Using the definition $\sinh(x) = \frac{e^x - e^{-x}}{2} \implies 2\sinh(x) = e^x - e^{-x}$, let $x = \frac{\hbar\omega}{2k_B T}$:$$C_V = k_B \left(\frac{\hbar\omega}{k_B T}\right)^2 \frac{1}{\left(2\sinh\left(\frac{\hbar\omega}{2k_B T}\right)\right)^2} = k_B \left(\frac{\hbar\omega}{2k_B T}\right)^2 \frac{1}{\sinh^2\left(\frac{\hbar\omega}{2k_B T}\right)}$$

#5.
Asymptotic Behavior (Discussion)In the high-temperature limit where $k_B T \gg \hbar\omega$, the parameter $x = \frac{\hbar\omega}{2k_B T} \to 0$. We evaluate using the Taylor series expansion $\sinh(x) \approx x$ for $x \to 0$:$$C_V \approx k_B (x)^2 \cdot \frac{1}{(x)^2} = k_B$$Conclusion: Yes, it matches the classical equipartition theorem prediction exactly. A classical 1D harmonic oscillator has two quadratic degrees of freedom (kinetic energy $p^2/2m$ and potential energy $\frac{1}{2}m\omega^2 q^2$), each contributing $\frac{1}{2}k_B$ to the heat capacity, totaling $C_V = k_B$.

##The Maximum Entropy Distribution under Mean and Variance ConstraintsWe want to find the continuous probability density function $p(x)$ maximizing the Shannon-Jaynes entropy functional with a uniform reference measure $m(x)=1$:$$\max_{p} H_c[p] = -\int_{-\infty}^{\infty} p(x) \ln p(x) dx$$Subject to three integral constraints:$\int_{-\infty}^{\infty} p(x) dx = 1$ (Normalization)$\int_{-\infty}^{\infty} x p(x) dx = \mu$ (Mean)$\int_{-\infty}^{\infty} (x-\mu)^2 p(x) dx = \sigma^2$ (Variance)We set up the functional Lagrangian:$$L[p] = -\int_{-\infty}^{\infty} p(x) \ln p(x) dx - \gamma \left( \int_{-\infty}^{\infty} p(x) dx - 1 \right) - \lambda_1 \left( \int_{-\infty}^{\infty} x p(x) dx - \mu \right) - \lambda_2 \left( \int_{-\infty}^{\infty} (x-\mu)^2 p(x) dx - \sigma^2 \right)$$Taking the functional derivative $\frac{\delta L}{\delta p(x)} = 0$:$$-\ln p(x) - 1 - \gamma - \lambda_1 x - \lambda_2 (x-\mu)^2 = 0$$$$p^*(x) = \exp(-(1+\gamma)) \exp\left(-\lambda_1 x - \lambda_2 (x-\mu)^2\right)$$Let $Z = \exp(1+\gamma)$. To preserve the localized expectation centering at $\mu$, the linear skew multiplier must vanish ($\lambda_1 = 0$). Thus:$$p^*(x) = \frac{1}{Z} \exp\left(-\lambda_2 (x-\mu)^2\right)$$We evaluate the parameters using standard Gaussian integrals ($\int_{-\infty}^{\infty} e^{-a(x-b)^2}dx = \sqrt{\frac{\pi}{a}}$):Normalization: $Z = \int_{-\infty}^{\infty} \exp\left(-\lambda_2 (x-\mu)^2\right) dx = \sqrt{\frac{\pi}{\lambda_2}}$Variance Match: $\sigma^2 = \int_{-\infty}^{\infty} (x-\mu)^2 \left( \sqrt{\frac{\lambda_2}{\pi}} \exp\left(-\lambda_2 (x-\mu)^2\right) \right) dx = \frac{1}{2\lambda_2} \implies \lambda_2 = \frac{1}{2\sigma^2}$Substituting $\lambda_2$ back into $Z$:$$Z = \sqrt{\frac{\pi}{1/(2\sigma^2)}} = \sqrt{2\pi\sigma^2}$$Reassembling the final PDF yields the classical Normal distribution:$$p^*(x) = \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left( -\frac{(x-\mu)^2}{2\sigma^2} \right) \implies X \sim \mathcal{N}(\mu, \sigma^2)$$

#Q3 Multiclass Logistic Regression via Maximum Entropy
## Automatic Classification of Inbound Customer Support Emails

This notebook solves for the optimal risk-neutral routing probabilities of an automated customer support routing agent.
We have two target destinations ($Y=0$: Finance, $Y=1$: Technical) and an observation state vector $x = (x_1, x_2, x_3)^T$.

The probability of routing to Technical support ($Y=1$) given an observation vector $x$ takes the logistic form:
$$P(Y=1|x) = \frac{\exp(\lambda_1 x_1 + \lambda_2 x_2 + \lambda_3 x_3)}{1 + \exp(\lambda_1 x_1 + \lambda_2 x_2 + \lambda_3 x_3)}$$

We optimize the parameters $\lambda_1, \lambda_2, \lambda_3$ to satisfy the observed joint historical expectations:
1. $E[x_1 \cdot I(Y=1)] = 0.05$
2. $E[x_2 \cdot I(Y=1)] = 0.15$
3. $E[x_3 \cdot I(Y=1)] = 0.16$

```python
import numpy as np
import pandas as pd
from scipy.optimize import minimize

# 1. Define the operational scenarios and their marginal probabilities P(x)
scenarios = {
    'A (Pure Financial)':    {'x': np.array([1, 0, 0]), 'P_x': 0.40},
    'B (Pure Access)':       {'x': np.array([0, 1, 0]), 'P_x': 0.25},
    'C (Pure Technical)':    {'x': np.array([0, 0, 1]), 'P_x': 0.05},
    'D (Access Technical)':  {'x': np.array([0, 1, 1]), 'P_x': 0.15},
    'E (Billing Access)':    {'x': np.array([1, 1, 0]), 'P_x': 0.05}
}

# 2. Historical target expectations E[x_i * I(Y=1)]
target_expectations = np.array([0.05, 0.15, 0.16])

print("Problem parameters successfully initialized.")
```

```python
def compute_conditional_prob(x, lambdas):
    """Computes P(Y=1 | x) using the logistic function."""
    dot_product = np.dot(lambdas, x)
    # Using stable sigmoid evaluation to prevent overflow
    return 1.0 / (1.0 + np.exp(-dot_product))

def loss_function(lambdas):
    """Computes the sum of squared errors between simulated and target expectations."""
    sim_expectations = np.zeros(3)

    # Calculate E[x_i * I(Y=1)] = Sum over all scenarios of [ x_i * P(Y=1|x) * P(x) ]
    for name, data in scenarios.items():
        x = data['x']
        P_x = data['P_x']
        P_Y1_given_x = compute_conditional_prob(x, lambdas)

        sim_expectations += x * P_Y1_given_x * P_x

    # Total Squared Error
    error = np.sum((sim_expectations - target_expectations) ** 2)
    return error

# Initial guess for the Lagrange multipliers [lambda_1, lambda_2, lambda_3]
initial_guess = [0.0, 0.0, 0.0]

# Run optimization
print("Optimizing parameters via Nelder-Mead...")
result = minimize(loss_function, initial_guess, method='Nelder-Mead', options={'xatol': 1e-8, 'maxiter': 2000})

if result.success:
    lambdas_opt = result.x
    print("\nOptimization Successful!")
    print(f"λ1 (Financial Weight): {lambdas_opt[0]:.4f}")
    print(f"λ2 (Access Weight):    {lambdas_opt[1]:.4f}")
    print(f"λ3 (Technical Weight): {lambdas_opt[2]:.4f}")
else:
    print("Optimization failed to converge.")
```

```python
# 1. Verify Expectations
final_expectations = np.zeros(3)
results_table = []

for name, data in scenarios.items():
    x = data['x']
    P_x = data['P_x']
    P_Y1_given_x = compute_conditional_prob(x, lambdas_opt)
    P_Y0_given_x = 1.0 - P_Y1_given_x

    final_expectations += x * P_Y1_given_x * P_x

    results_table.append({
        'Scenario': name,
        'State Vector (x)': str(list(x)),
        'P(x)': P_x,
        'P(Technical | x)': f"{P_Y1_given_x * 100:.2f}%",
        'P(Finance | x)': f"{P_Y0_given_x * 100:.2f}%"
    })

print("=== Constraint Verification ===")
for i in range(3):
    print(f"Constraint {i+1} (x{i+1}): Target = {target_expectations[i]:.4f} | Inferred Model = {final_expectations[i]:.4f}")

print("\n=== Scenario Routing Summary ===")
df = pd.DataFrame(results_table)
from IPython.display import display
display(df)
```

# Asset Pricing via Maximum Entropy Principle

This notebook implements the Maximum Entropy Principle (MaxEnt) to recover the risk-neutral probability density function $p(S_T)$ of a future asset price $S_T$, subject to observed market constraints:
1. $E[S_T] = 102$ (Forward price)
2. $E[S_T^2] = 10600$ (Volatility/Variance related)
3. $E[\ln(S_T)] = 4.60$ (Skewness related)

The theoretical density is given by:
$$p(S_T) = \frac{1}{Z} S_T^{-\lambda_3} \exp(-\lambda_1 S_T - \lambda_2 S_T^2)$$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad
from scipy.optimize import minimize

# Target market constraints
E_S = 102.0
E_S2 = 10600.0
E_logS = 4.60

print("Dependencies loaded successfully. Target constraints initialized.")
```

```python
def get_density_and_Z(S, l1, l2, l3):
    """Calculates the unnormalized density piece for integration safety."""
    # Prevent overflow/underflow issues by clipping values if necessary
    exponent = -l1 * S - l2 * (S**2)
    # Avoid log(0) issues by adding a tiny epsilon if S is an array containing 0
    S_term = np.power(S, -l3)
    return S_term * np.exp(exponent)

def loss_function(lambdas):
    """
    Dual objective function or residual function to minimize.
    We minimize the squared error of our constraints to find the correct lambdas.
    """
    l1, l2, l3 = lambdas

    # Define the integrand functions for Z and expectations
    # We integrate from a tiny positive number to a large upper bound to approximate infinity safely
    lower_bound = 1e-5
    upper_bound = 500.0

    Z, _ = quad(lambda s: get_density_and_Z(s, l1, l2, l3), lower_bound, upper_bound)

    if Z <= 0 or np.isnan(Z) or np.isinf(Z):
        return 1e10  # Return a massive penalty if integration fails

    # Calculate expected values
    val_S, _ = quad(lambda s: s * get_density_and_Z(s, l1, l2, l3), lower_bound, upper_bound)
    val_S2, _ = quad(lambda s: (s**2) * get_density_and_Z(s, l1, l2, l3), lower_bound, upper_bound)
    val_logS, _ = quad(lambda s: np.log(s) * get_density_and_Z(s, l1, l2, l3), lower_bound, upper_bound)

    sim_E_S = val_S / Z
    sim_E_S2 = val_S2 / Z
    sim_E_logS = val_logS / Z

    # Compute squared residuals
    error = (sim_E_S - E_S)**2 + (sim_E_S2 - E_S2)**2 + (sim_E_logS - E_logS)**2
    return error

# Initial guess for [lambda_1, lambda_2, lambda_3]
# Note: lambda_2 must be positive to ensure the tail of the distribution converges to 0.
initial_guess = [0.01, 0.0001, -0.5]

print("Optimizing Lagrange Multipliers... This might take a few seconds.")
res = minimize(loss_function, initial_guess, method='Nelder-Mead', options={'xatol': 1e-6, 'maxiter': 2000})

if res.success:
    l1_opt, l2_opt, l3_opt = res.x
    print("\nOptimization Successful!")
    print(f"λ1 = {l1_opt:.6f}")
    print(f"λ2 = {l2_opt:.6f}")
    print(f"λ3 = {l3_opt:.6f}")
else:
    print("Optimization failed to converge perfectly. Try adjusting the initial guess.")
```

```python
# Compute final Partition Function Z with optimized values
lower_bound = 1e-5
upper_bound = 500.0
Z_opt, _ = quad(lambda s: get_density_and_Z(s, l1_opt, l2_opt, l3_opt), lower_bound, upper_bound)

# Verify constraints
final_E_S, _ = quad(lambda s: s * get_density_and_Z(s, l1_opt, l2_opt, l3_opt), lower_bound, upper_bound)
final_E_S2, _ = quad(lambda s: (s**2) * get_density_and_Z(s, l1_opt, l2_opt, l3_opt), lower_bound, upper_bound)
final_E_logS, _ = quad(lambda s: np.log(s) * get_density_and_Z(s, l1_opt, l2_opt, l3_opt), lower_bound, upper_bound)

print("--- Verification ---")
print(f"Target E[S]   = {E_S} | Model E[S]   = {final_E_S/Z_opt:.2f}")
print(f"Target E[S^2] = {E_S2} | Model E[S^2] = {final_E_S2/Z_opt:.2f}")
print(f"Target E[lnS] = {E_logS}  | Model E[lnS] = {final_E_logS/Z_opt:.2f}")

# Generate values for plotting
S_vals = np.linspace(50, 160, 500)
p_vals = get_density_and_Z(S_vals, l1_opt, l2_opt, l3_opt) / Z_opt

# Plot the distribution
plt.figure(figsize=(10, 5))
plt.plot(S_vals, p_vals, label='MaxEnt Density $p(S_T)$', color='darkblue', lw=2)
plt.axvline(E_S, color='red', linestyle='--', label=f'Forward Price (E[S] = {E_S})')
plt.title('Maximum Entropy Risk-Neutral Probability Density', fontsize=14)
plt.xlabel('Future Asset Price ($S_T$)', fontsize=12)
plt.ylabel('Probability Density', fontsize=12)
plt.grid(True, alpha=0.3)
plt.legend(fontsize=11)
plt.show()
```
