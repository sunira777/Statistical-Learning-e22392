#Q1) Prediction of fraudulent transactions

A bank monitors credit-card transactions for possible fraud.

For each transaction, let the sample space be

$$
\Omega=\{\omega_1,\omega_2,\omega_3,\omega_4\},
$$

where:

* $\omega_1$: low-risk customer, legitimate transaction,
* $\omega_2$: low-risk customer, fraudulent transaction,
* $\omega_3$: high-risk customer, legitimate transaction,
* $\omega_4$: high-risk customer, fraudulent transaction.

Assume the probabilities are

$$
P(\{\omega_1\})=0.50,\qquad
P(\{\omega_2\})=0.10,\qquad
P(\{\omega_3\})=0.20,\qquad
P(\{\omega_4\})=0.20.
$$

Let $X:\Omega\to\mathbb R$ denote the fraud indicator:

$$
X(\omega_1)=0,\qquad
X(\omega_2)=1,\qquad
X(\omega_3)=0,\qquad
X(\omega_4)=1.
$$

Thus $X=1$ means the transaction is fraudulent, and $X=0$ means it is legitimate.

Now suppose the bank does not observe the full outcome $\omega$, but only the customer's risk class. Define $Y:\Omega\to\mathbb R$ by

$$
Y(\omega_1)=0,\qquad
Y(\omega_2)=0,\qquad
Y(\omega_3)=1,\qquad
Y(\omega_4)=1.
$$

Thus:

* $Y=0$ means “low-risk customer,”
* $Y=1$ means “high-risk customer.”

Answer the following.

1. Write down the probability space $(\Omega,\mathfrak F,P)$, where $\mathfrak F=\mathfrak P(\Omega)$.

2. Compute the $\sigma$-algebra generated by $Y$, namely $\sigma(Y)$.

3. Explain why $E[X\mid Y]$ must be $\sigma(Y)$-measurable, and therefore must be constant on the sets $\{ \omega_1,\omega_2\}$ and $\{\omega_3,\omega_4\}$.

4. Compute $E[X\mid Y]$ explicitly and show that

$$
E[X\mid Y]= \frac{1}{6} \mathbb{1}_{\{Y=0\}} + \frac{1}{2} \mathbb{1}_{\{Y=1\}}
$$

5. Verify the defining property of conditional expectation by checking that for every set $A\in\sigma(Y)$,

$$
\int_A E[X\mid Y]\,dP=\int_A X\,dP.
$$

6. Find the Conditional PMF $p_{X|Y}(x \mid y)$.

7. Verify the identity:
$$\mathbb{E}[X\mid Y=y] = \sum_{x \in \{0,1\}} x \cdot p_{X|Y}(x \mid y)$$.

8. Calculate the **Mean Square Error (MSE)**

9. How would the **Mean Square Error (MSE)** change if the bank decided to ignore the customer risk class $Y$ and instead used the global average fraud rate ($E[X] = 0.30$) as a constant estimate for every transaction? Calculate this error and explain its relationship to the **Law of Total Variance**.

10. Using the joint distribution of $(X, Y)$ derived from the sample space:

    a) Compute the entropy $H(X)$, representing the bank's initial uncertainty regarding fraud.

    b) Compute the conditional entropy $H(X \mid Y)$, representing the uncertainty remaining after the customer's risk class is revealed.

    c) Calculate the Mutual Information $I(X; Y)$.

    d) Based on these values, what percentage of the total uncertainty about fraud is removed by knowing the customer's risk class?

## [8]

We compute the Minimum Mean Square Error (MMSE), which represents the smallest possible mean squared error when using the optimal estimator \(\mathbb{E}[X \mid Y]\). It is given by

$$
\text{MSE}_{\min} = \mathbb{E}\big[(X - \mathbb{E}[X \mid Y])^2\big] = \mathbb{E}[\mathrm{Var}(X \mid Y)].
$$

---

### Step 1: Conditional variances

Since \(X\) is Bernoulli given \(Y=y\), we use \(\mathrm{Var}(X \mid Y=y) = p(1-p)\).

For \(Y=0\):

$$
\mathrm{Var}(X \mid Y=0)
= \frac{1}{6}\left(1-\frac{1}{6}\right)
= \frac{1}{6}\cdot\frac{5}{6}
= \frac{5}{36}
$$

For \(Y=1\):

$$
\mathrm{Var}(X \mid Y=1)
= \frac{1}{2}\left(1-\frac{1}{2}\right)
= \frac{1}{2}\cdot\frac{1}{2}
= \frac{1}{4}
$$

---

### Step 2: Take expectation over \(Y\)

$$
\mathbb{E}[\mathrm{Var}(X \mid Y)]
= \mathrm{Var}(X \mid Y=0)P(Y=0)
+ \mathrm{Var}(X \mid Y=1)P(Y=1)
$$

Substituting values:

$$
= \left(\frac{5}{36}\cdot 0.60\right)
+ \left(\frac{1}{4}\cdot 0.40\right)
$$

$$
= \left(\frac{5}{36}\cdot \frac{3}{5}\right)
+ \left(\frac{1}{4}\cdot \frac{2}{5}\right)
$$

$$
= \frac{1}{12} + \frac{1}{10}
= \frac{11}{60}
\approx 0.1833
$$

---

### Final result

$$
\text{MMSE} = \frac{11}{60} \approx 0.1833
$$

This is the lowest possible average squared error achievable when estimating \(X\) from \(Y\).

# [9]

When we use a constant estimator \(c = \mathbb{E}[X]\), the mean square error becomes the variance of \(X\):

$$
\text{MSE}_{\text{global}} = \mathbb{E}\big[(X - \mathbb{E}[X])^2\big] = \mathrm{Var}(X)
$$

Since \(X\) is Bernoulli with \(P(X=1)=0.30\),

$$
\mathrm{Var}(X) = p(1-p) = 0.30 \times 0.70 = 0.21
$$

---

## Comparison of Errors

- **Optimal error (using \(Y\))**
$$
\text{MSE}_{\min} = \mathbb{E}[\mathrm{Var}(X \mid Y)] = \frac{11}{60} \approx 0.1833
$$

- **Global error (without \(Y\))**
$$
\mathrm{Var}(X) = 0.21
$$

So, using the information in \(Y\) reduces the prediction error from **0.21** to **0.1833**.

---

## Law of Total Variance

The total variance can be decomposed as:

$$
\mathrm{Var}(X) = \mathbb{E}[\mathrm{Var}(X \mid Y)] + \mathrm{Var}(\mathbb{E}[X \mid Y])
$$

In this example:

- **Unexplained part:**
$$
\mathbb{E}[\mathrm{Var}(X \mid Y)] = 0.1833
$$

- **Explained part:**
$$
\mathrm{Var}(\mathbb{E}[X \mid Y]) = 0.21 - 0.1833 = 0.0267
$$

---

## Final decomposition

$$
0.21 = 0.1833 + 0.0267
$$

---

## Interpretation

The first term represents the irreducible uncertainty even after using \(Y\), while the second term shows how much uncertainty is reduced by using \(Y\). This difference measures the predictive value of \(Y\).

# Q2) Conditional Density of the Multivariate Gaussian

Let $\mathbf{Z}$ be the concatenated random matrix
$\mathbf{Z} = \begin{bmatrix} \mathbf{X} \\ \mathbf{Y} \end{bmatrix} \in \mathbb{R}^{n+m}$.


The random variable $\mathbf{Z}$ is said to have a **joint multivariate normal distribution** if every linear combination of its components follows a univariate normal distribution.

For $\mathbf{Z}$ to possess a **joint density function** (meaning its distribution is absolutely continuous with respect to the Lebesgue measure on $\mathbb{R}^{n+m}$), its covariance matrix must be non-singular (positive definite).


The joint density function $f_{\mathbf{X}, \mathbf{Y}}(\mathbf{z})$ for $\mathbf{z} \in \mathbb{R}^{n+m}$ is given by:

$$f_{\mathbf{X}, \mathbf{Y}}(\mathbf{z}) = \frac{1}{\sqrt{(2\pi)^{n+m} |\boldsymbol{\Sigma}|}} \exp\left( -\frac{1}{2} (\mathbf{z} - \boldsymbol{\mu})^T \boldsymbol{\Sigma}^{-1} (\mathbf{z} - \boldsymbol{\mu}) \right).$$

Here:

* **Mean ($\boldsymbol{\mu}$):** $\boldsymbol{\mu} = E[\mathbf{Z}] = \begin{bmatrix} \boldsymbol{\mu}_X \\ \boldsymbol{\mu}_Y \end{bmatrix} \in \mathbb{R}^{n+m}$, where $\boldsymbol{\mu}_X \in \mathbb{R}^n$ and $\boldsymbol{\mu}_Y \in \mathbb{R}^m$.
* **Covariance Matrix ($\boldsymbol{\Sigma}$):** A symmetric, positive definite matrix $\boldsymbol{\Sigma} \in \mathbb{R}^{(n+m) \times (n+m)}$ partitioned as:
$$\boldsymbol{\Sigma} = \begin{bmatrix} \boldsymbol{\Sigma}_{XX} & \boldsymbol{\Sigma}_{XY} \\ \boldsymbol{\Sigma}_{YX} & \boldsymbol{\Sigma}_{YY} \end{bmatrix}$$
    * $\boldsymbol{\Sigma}_{XX} = \text{Var}(\mathbf{X})$ (size $n \times n$)
    * $\boldsymbol{\Sigma}_{YY} = \text{Var}(\mathbf{Y})$ (size $m \times m$)
    * $\boldsymbol{\Sigma}_{XY} = \text{Cov}(\mathbf{X}, \mathbf{Y}) = \boldsymbol{\Sigma}_{YX}^T$ (size $n \times m$)
* **Determinant ($|\boldsymbol{\Sigma}|$):** The normalization factor ensuring the total integral is 1.

Show that the conditional distribution is given by $f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y})$

$$f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y}) = \frac{1}{\sqrt{(2\pi)^n |\mathbf{S}_{XX}|}} \exp\left( -\frac{1}{2} (\mathbf{x} - \boldsymbol{\mu}^*)^T \boldsymbol{\Sigma}_{X\mid Y}^{-1} (\mathbf{x} - \boldsymbol{\mu}^*) \right),$$
where
* $\boldsymbol{\Sigma}_{X\mid Y} = \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}$
* $\boldsymbol{\mu}^* = \boldsymbol{\mu}_x + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y} - \boldsymbol{\mu}_y)$

# Answer for Q2

To derive the conditional density $f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y})$, we use the definition of conditional probability for densities:

$$f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y}) = \frac{f_{\mathbf{X}, \mathbf{Y}}(\mathbf{x}, \mathbf{y})}{f_{\mathbf{Y}}(\mathbf{y})}$$

Since we have already decomposed the joint density and the marginal density of $\mathbf{Y}$ in the previous steps, this becomes a task of algebraic subtraction in the exponent.

---

From our previous derivation, we factored the joint quadratic form $Q$ using the Schur complement of $\boldsymbol{\Sigma}_{YY}$:
$$Q_{joint} = (\mathbf{y} - \boldsymbol{\mu}_y)^T \boldsymbol{\Sigma}_{YY}^{-1} (\mathbf{y} - \boldsymbol{\mu}_y) + (\mathbf{x} - \boldsymbol{\mu}^*)^T \mathbf{S}_{XX}^{-1} (\mathbf{x} - \boldsymbol{\mu}^*)$$

Where:
* $\mathbf{S}_{XX} = \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}$
* $\boldsymbol{\mu}^* = \boldsymbol{\mu}_x + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y} - \boldsymbol{\mu}_y)$

The marginal density $f_{\mathbf{Y}}(\mathbf{y})$ has the exponent:
$$Q_{marginal} = (\mathbf{y} - \boldsymbol{\mu}_y)^T \boldsymbol{\Sigma}_{YY}^{-1} (\mathbf{y} - \boldsymbol{\mu}_y)$$

When we divide the densities, we subtract the exponents ($e^{A} / e^{B} = e^{A-B}$):
$$Q_{conditional} = Q_{joint} - Q_{marginal} = (\mathbf{x} - \boldsymbol{\mu}^*)^T \mathbf{S}_{XX}^{-1} (\mathbf{x} - \boldsymbol{\mu}^*)$$

---

Now we look at the coefficients.
* **Joint constant:** $\frac{1}{\sqrt{(2\pi)^{n+m} |\boldsymbol{\Sigma}|}}$
* **Marginal constant:** $\frac{1}{\sqrt{(2\pi)^m |\boldsymbol{\Sigma}_{YY}|}}$

Dividing the joint constant by the marginal constant gives:
$$\frac{1}{\sqrt{(2\pi)^{n+m} |\boldsymbol{\Sigma}|}} \cdot \sqrt{(2\pi)^m |\boldsymbol{\Sigma}_{YY}|} = \frac{1}{\sqrt{(2\pi)^n \frac{|\boldsymbol{\Sigma}|}{|\boldsymbol{\Sigma}_{YY}|}}}$$

Using the determinant identity $|\boldsymbol{\Sigma}| = |\boldsymbol{\Sigma}_{YY}| \cdot |\mathbf{S}_{XX}|$, the term $\frac{|\boldsymbol{\Sigma}|}{|\boldsymbol{\Sigma}_{YY}|}$ simplifies exactly to $|\mathbf{S}_{XX}|$.

---

Combining the simplified constant and the remaining quadratic form, we get:

$$f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y}) = \frac{1}{\sqrt{(2\pi)^n |\mathbf{S}_{XX}|}} \exp\left( -\frac{1}{2} (\mathbf{x} - \boldsymbol{\mu}^*)^T \mathbf{S}_{XX}^{-1} (\mathbf{x} - \boldsymbol{\mu}^*) \right)$$

This is the density of a multivariate normal distribution:
$$(\mathbf{X} \mid \mathbf{Y} = \mathbf{y}) \sim \mathscr{N}(\boldsymbol{\mu}_x + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y} - \boldsymbol{\mu}_y), \quad \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}).$$

---

**Summary of the Parameters**
* **Conditional Mean:** $\boldsymbol{\mu}_{X|Y} = \boldsymbol{\mu}_x + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y} - \boldsymbol{\mu}_y)$
    * *Insight:* The mean is adjusted by how much $\mathbf{y}$ deviates from its expected value, scaled by the "regression" coefficient $\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}$.
* **Conditional Covariance:** $\boldsymbol{\Sigma}_{X|Y} = \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}$
    * *Insight:* Notice this does **not** depend on the specific value of $\mathbf{y}$. The variance is always reduced (or stays the same) compared to the marginal $\boldsymbol{\Sigma}_{XX}$.

Does the fact that the conditional variance is independent of the observed value $y$ spark any ideas regarding its application in linear regression or Kalman filters?

## 1. Joint Distribution

We are given:
X ~ N(0,1),  Q ~ N(0,4),  Y = X + Q

Since X and Q are independent normal variables, (X, Y) is jointly normal.

---

### Distribution of Y

Var(Y) = Var(X) + Var(Q) = 1 + 4 = 5

So,
Y ~ N(0, 5)

---

### Covariance

Cov(X, Y) = Cov(X, X + Q)
           = Var(X) + 0
           = 1

So,
Cov(X, Y) = 1

---

## 2. Why X is not determined by Y

Since Y = X + Q includes noise Q, the same value of Y can come from many different pairs (X, Q).

So X cannot be uniquely determined from Y.

Therefore, X is not measurable from Y.

---

## 3. Conditional Expectation E[X | Y]

For jointly normal variables:

E[X | Y=y] = E[X] + Cov(X,Y)/Var(Y) * (y - E[Y])

Substitute values:

E[X | Y=y] = (1/5) y

So,
E[X | Y] = (1/5) Y

---

## 4. Error (Residual)

Let:
Z = X - E[X | Y] = X - (1/5)Y

### Mean

E[Z] = 0

---

### Orthogonality

Cov(Z, Y) = Cov(X - (1/5)Y, Y)

           = Cov(X,Y) - (1/5)Var(Y)

           = 1 - (1/5)*5 = 0

So Z is uncorrelated with Y.

---

## 5. Differential Entropy

h(X) = (1/2) log(2 * pi * e * 1)

---

Conditional variance:

Var(X | Y) = 1 - (1^2 / 5) = 4/5

So,

h(X | Y) = (1/2) log(2 * pi * e * (4/5))

---

## 6. Mutual Information

I(X;Y) = h(X) - h(X | Y)

        = (1/2) log(5/4)

---

### Effect of noise

If Var(Q) becomes very large:
- Y becomes noisy
- Cov(X,Y)/Var(Y) becomes small
- I(X;Y) goes to 0

So Y gives almost no information about X.

---


##7. Diagnostic Interpretation

The manager's suggestion to use $X = Y$ is a **Naive Estimator**.
* If we use $X = Y$, our error variance would be $E[(X-Y)^2] = E[Q^2] = 4$.
* By using $E[X|Y] = 0.2Y$, our error variance is only $0.8$.

**Shrinkage:** The factor $0.2$ is called a shrinkage estimator. It tells us that when a sensor is noisy, the "best" guess is to be conservative and pull the reading back toward the historical average (zero). If the power spikes, we don't immediately assume the tool is melting; we assume most of that spike is just the material being tough.
